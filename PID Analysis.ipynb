```python
from scipy.signal import *
from IPython.html.widgets import *
from matplotlib.pyplot import *
from numpy import *
import control as c
%matplotlib inline
rcParams['figure.figsize']=(8,4)
```

```python
G=c.tf(10,[1,6,11,16])
m = c.step(G)
plot(m[1],m[0])
grid(1)
title("Step response of Open-Loop System")
xlabel('Time')
```

```python
def plotPIDres(Kp,Ki,Kd):
    G=c.tf(10,[1,6,11,16])
    Gc = c.tf([Kd,Kp,Ki],[1,0])
    M_P= c.feedback(Gc*G,1)
    y_P = c.step(M_P,linspace(0,5,100))
    plot(y_P[1],y_P[0])
    grid(1)
    xlabel("Time")
    ylim([0,2])
    
```

```python
interact(plotPIDres,Kp=(0,20,.5),Ki=(0,20,.5),Kd=(0,20,.5))
```
